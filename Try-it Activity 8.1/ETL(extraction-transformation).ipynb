The excel file is read

```python
worksheet = pd.read_excel('data/mrtssales92-present.xls', sheet_name=None, skiprows=4, nrows=66)
```

This excel file contains many sheets that have to be transformed. In order to do so, the excel file has to be iterated as a dictionary, where the keys are the sheets and the values are the dataset of each sheet.

The first step is renaming the columns of the NAICS ID and the Business description. Also, the NaN values from the NAICS columns are dropped.

```python
for sheetname, df in worksheet.items():
    df.rename(columns={'Unnamed: 0': 'NAICS', 'Unnamed: 1': 'Business Description'}, inplace=True)
    df = df.dropna(subset=['NAICS'], inplace=True)
worksheet['2005']
```

The index is reset. 

```python
for sheetname, df in worksheet.items():
    df.reset_index(drop=True, inplace=True)
worksheet['2021'].head()
```

The format of the dates are adjusted and homogenized for each and every sheet.

```python
for sheetname, df in worksheet.items():
    if sheetname == '2021':
        df.rename(columns={"Feb. 2021(p)":"Feb. 2021"}, inplace=True)  
    else:   
        df.rename(columns={df.columns[6]:"May. " + sheetname}, inplace=True)
worksheet['2021'].head()
```

```python
for sheetname, df in worksheet.items():
    if sheetname == '2021':
        n = 3
        while n > 1:
            df.rename(columns={df.columns[n]:df.columns[n].replace(".", "")}, inplace=True)
            n = n - 1
    else:
        n = 13
        while n > 1:
            df.rename(columns={df.columns[n]:df.columns[n].replace(".", "")}, inplace=True)
            n = n - 1
worksheet['2005'].head()
```

The dates are transformed into a date type structure.

```python
from datetime import datetime as dt

for sheetname, df in worksheet.items():
    if sheetname == '2021':
        n = 3
        while n > 1:
            df.rename(columns={df.columns[n]:pd.to_datetime(df.columns[n], format='%b %Y').date().strftime("%Y/%m/%d")}, inplace=True)
            n = n - 1
    else:
        n = 13
        while n > 1:
            df.rename(columns={df.columns[n]:pd.to_datetime(df.columns[n], format='%b %Y').date().strftime("%Y/%m/%d")}, inplace=True)
            n = n - 1
worksheet['2005'].head()
```

The "Total", "CY CUM" and "PY CUM" columns are dropped in each sheet.

```python
for sheetname, df in worksheet.items():
        df.drop(columns=[col for col in ["TOTAL", "CY CUM", "PY CUM"] if col in df.columns], inplace=True)
worksheet['2005'].head()
```

The sheets are iterated to create a new dataframe, which will contain all the values in one page. This is to create the table in the database containing all information.

```python
list = []
for sheetname, df in worksheet.items():
    if sheetname == "2021":
        y = 1
        while y < 3:
            i=0
            while i < 58:
                temp_naics = df.iloc[i,0]
                temp_description = df.iloc[i,1]
                temp_date = df.columns[y+1]
                temp_sales = df.iloc[i,y+1]
                dict1 = {'NAICS_ID': temp_naics, 'Business Description': temp_description, 'Date': temp_date, "Sales":temp_sales}
                i = i+1
                list.append(dict1)
            y = y + 1
    else:
        y = 1
        while y < 13:
            i=0
            while i < 58:
                temp_naics = df.iloc[i,0]
                temp_description = df.iloc[i,1]
                temp_date = df.columns[y+1]
                temp_sales = df.iloc[i,y+1]
                dict1 = {'NAICS_ID': temp_naics, 'Business Description': temp_description, 'Date': temp_date, "Sales":temp_sales}
                i = i+1
                list.append(dict1)
            y = y + 1 
df_final = pd.DataFrame(list)
df_final
```

The "NAICS_ID" table gets a new format.

```python
df_final['NAICS_ID'] = df_final['NAICS_ID'].replace(",","-", regex=True)
df_final['NAICS_ID'] = df_final['NAICS_ID'].replace(" ", "", regex=True)
df_final
```

The new dataframe gets a new column indicating the status of the row, and the "Sales" column gets a new format to be loaded as an *int* type into the database.

```python
df_final['Status'] = "Estimated"
df_final
```

```python
df_final['Status'].where(df_final.Sales != "(S)", "Suppressed", inplace=True)
df_final
```

```python
df_final['Sales'].replace("(S)", 0, regex=True, inplace=True)
df_final.iloc[11]
```

```python
import numpy as np


df_final['Status'].where(df_final.Sales != "(NA)", "NaN", inplace=True)
df_final['Sales'].replace("(NA)", None, regex=True, inplace=True)
```

The "NAICS_ID" column is saved as a *String* type. A new "ID_GROUP" column is created from the "NAICS_ID" one.

```python
df_final['NAICS_ID'] = df_final['NAICS_ID'].astype(str)
df_final['ID_GROUP'] = df_final.loc[:,'NAICS_ID']
df_final
```

```python
df_final['ID_GROUP'] = df_final['ID_GROUP'].str.slice(start=0, stop=3)
df_final
```

The dataframe is reindexed.  

```python
df_final = df_final.reindex(columns=['NAICS_ID', 'ID_GROUP', 'Business Description', 'Date', 'Sales', 'Status'])
df_final
```

Finally, the new dataframe is turned into a CSV file, which will be uploaded into the database in another Python script.

```python
df_final.to_csv('data/transformedmrtsales92-present.csv', index=False)
```
